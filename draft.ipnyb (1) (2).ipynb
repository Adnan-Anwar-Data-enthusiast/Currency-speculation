```python
def split_swap_columns(df):
    split_dataframes = {}

    for i in range(0, len(df.columns), 2):
        date_col = df.columns[i]
        value_col = df.columns[i + 1]

        # Extract the swap name and time period from the value_col
        try:
            parts = value_col.split('_')
            swap = parts[0]
            period = parts[-1]
            key = f"{swap}_{period}"
        except IndexError:
            continue

        split_dataframes[key] = df[[date_col, value_col]].copy()

    return split_dataframes
```
